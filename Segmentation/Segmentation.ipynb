```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("data.jpg",0)
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
blurred = cv2.GaussianBlur(img, (3, 3), 0)
'''blurred = cv2.medianBlur(img,3)'''
```

```python
plt.imshow(blurred)
cv2.imshow("",blurred)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret,thresh1 = cv2.threshold(blurred,127,255,cv2.THRESH_BINARY)
plt.imshow(thresh1)
cv2.imshow("",thresh1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
histr = cv2.calcHist([thresh1[10]],[0],None,[256],[0,256])
plt.plot(histr)
plt.show()
```

```python
whites=[]
blacks=[]
for i in range(img.shape[0]):
    black=np.unique(thresh1[i],return_counts=True)[1][0]
    white=img.shape[1]-black
    whites.append(white)
    blacks.append(black)
print(whites)
print(len(whites))
print(blacks)
```

```python
plt.plot(whites)
plt.show()
```

```python
wlist=[]
for i in range(len(whites)):
    if whites[i]>440:
        wlist.append(i)
```
