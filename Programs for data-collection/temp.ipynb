```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("img1.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 75, 150)
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 30, maxLineGap=20)
for line in lines:
   x1, y1, x2, y2 = line[0]
   cv2.line(img, (x1, y1), (x2, y2), (0, 0, 128), 1)
cv2.imshow("linesEdges", edges)
cv2.imshow("linesDetected", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray=img.copy()
gray = cv2.bitwise_not(gray)
thresh1=cv2.adaptiveThreshold(gray,1,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, 15, 2)
```

```python
plt.imshow(thresh1)
plt.show()
```

```python
thresh1=cv2.bitwise_not(thresh1)
plt.imshow(thresh1)
plt.show()
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
```

```python
eroded=cv2.erode(thresh1,kernel,iterations=1)
final=cv2.dilate(eroded,kernel,iterations=1)
```

```python
plt.imshow(final)
plt.show()
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(1,100))
final2=cv2.erode(final,kernel,iterations=1)
```

```python
plt.imshow(final2)
plt.show()
```

```python
edges = cv2.Canny(final, 75, 150)
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 100, maxLineGap=20)
for line in lines:
   x1, y1, x2, y2 = line[0]
   cv2.line(img, (x1, y1), (x2, y2), (0, 0, 128), 1)
cv2.imshow("linesEdges", edges)
cv2.imshow("linesDetected", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
